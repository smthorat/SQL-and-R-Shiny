```python
import pandas as pd
```

```python
data = pd.read_csv("/Volumes/Jagannath/Projects/SQL_Project /ncbi_dataset-2.tsv", sep='\t')
```

```python
data
```

## Handle missing values

```python

#remove rows where 'Gene Type' is missing
data = data.dropna(subset=['Gene Type'])
data 
```

```python
# Drop columns that have too many missing values
data = data.drop(columns=['Transcripts', 'Gene Group Identifier', 'Gene Group Method'])
data
```

## Standardize Text Fields

```python
# Convert 'Symbol' column to uppercase
data['Symbol'] = data['Symbol'].str.upper()
data['Symbol']
```

```python
# Convert 'Gene Type' to title case (first letter of each word capitalized)
data['Gene Type'] = data['Gene Type'].str.title()
data['Gene Type']
```

```python
# Display the column names in the DataFrame
print(data.columns)
```

```python
# Trim any leading or trailing whitespace in all string columns
data = data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
data
```

```python
# Display the cleaned data to verify the results
print("Cleaned Data Preview:")
print(data.head())
```

```python
# Summary statistics to confirm data cleaning
print("\nData Summary After Cleaning:")
print(data.info())
```

```python

# Save the cleaned data to a new file (optional)
cleaned_file_path = '/Volumes/Jagannath/Projects/cleaned_gene_data.tsv'
data.to_csv(cleaned_file_path, sep='\t', index=False)
print(f"\nCleaned data has been saved to {cleaned_file_path}")
```
